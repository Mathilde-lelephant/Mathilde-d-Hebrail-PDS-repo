## Generador de señale senoidal (TP1)

Para este TP necesitamos un señal senoidal, entonces pongo aqui el trabajo del TP1 que nos daba este señal:


```python
# Módulos para Jupyter
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (10,10)
import matplotlib.pyplot as plt
import pdsmodulos as pds
import math
from math import *

#%% Esto tiene que ver con cuestiones de presentación de los gráficos,
# NO ES IMPORTANTE
fig_sz_x = 14
fig_sz_y = 13
fig_dpi = 80 # dpi

fig_font_size = 16

plt.rcParams.update({'font.size':fig_font_size})
##############################################
#%% A partir de aquí comienza lo IMPORTANTE #
#############################################

def mi_funcion_sen( vmax, dc, ff, ph, nn, fs ):
    
    ts = 1/fs # tiempo de muestreo
    df = fs/nn # resolución espectral
    
    # grilla de sampleo temporal
    tt = np.linspace(0, (N-1)*ts, N).flatten()
    
    # guardaremos las señales creadas al ir poblando la siguiente matriz vacía
    x = np.array([], dtype=np.float).reshape(N,0)
    
    aux = dc + vmax * np.sin( 2*np.pi*ff*tt + ph )
    xx = aux.reshape(N,1)
    
    
    return tt, xx
```

```python
    # Datos generales de la simulación
vmax = 1 #amplitud maximal de la senoidal (volts)
dc = 0 #valor medio (volts)
N = 2000   # cantidad de muestras
ff = N/4 #frequencial del señal (Hz)
ph = np.pi/2 #fase (radianes)
fs = N # frecuencia de muestreo (Hz)

    #Creation del señal
tt, xx = mi_funcion_sen( vmax, dc, ff, ph, N, fs)  
    
    # Presentación gráfica del señal
plt.figure(1)
line_hdls = plt.plot(tt, xx)
plt.title('Señal senoidal' )
plt.xlabel('tiempo [segundos]')
plt.ylabel('Amplitud [V]')
plt.grid(which='both', axis='both')
plt.show()
    
```

## Mi funcion DFT (TP2)

```python
def mi_funcion_DFT(xx):
    N = len(xx)
    xx_spectrum = xx * complex(0, 0) * 0
    for k in range(0, N):
        for n in range(0, N):
            xx_spectrum[k] = xx_spectrum[k] + xx[n] * (math.cos(2 * math.pi * k * n/N) + math.sin(2 * math.pi * k * n/N) * 1j)    
    return xx_spectrum

```

Testamos la funcion sobre el siñal que hemos creado:

```python
xx_spectrum = mi_funcion_DFT(xx)

plt.figure(2)
line_hdls = plt.plot((np.abs(xx_spectrum)))
plt.title('DFT del Señal senoidal' )
plt.xlabel('frequencia [Hs]')
plt.ylabel('Amplitud')
plt.grid(which='both', axis='both')
plt.show()
```

La DFT del señal senoidal nos da un dirac al lugar de la frequencia del señal (N/4 = 2000/4 = 500Hz), como esperado. Esto muestra que la funcion funciona bien.
